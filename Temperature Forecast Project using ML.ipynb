### Handling Missing Values:

```python
#Drop Rows with Missing Values
df.dropna(inplace=True)
```

```python
# Impute numerical columns with median
df['Present_Tmax'].fillna(df['Present_Tmax'].median(), inplace=True)
df['Present_Tmin'].fillna(df['Present_Tmin'].median(), inplace=True)
df['LDAPS_RHmin'].fillna(df['LDAPS_RHmin'].median(), inplace=True)
df['LDAPS_RHmax'].fillna(df['LDAPS_RHmax'].median(), inplace=True)
df['LDAPS_Tmax_lapse'].fillna(df['LDAPS_Tmax_lapse'].median(), inplace=True)
df['LDAPS_Tmin_lapse'].fillna(df['LDAPS_Tmin_lapse'].median(), inplace=True)
df['LDAPS_WS'].fillna(df['LDAPS_WS'].median(), inplace=True)
df['LDAPS_LH'].fillna(df['LDAPS_LH'].median(), inplace=True)
df['LDAPS_CC1'].fillna(df['LDAPS_CC1'].median(), inplace=True)
df['LDAPS_CC2'].fillna(df['LDAPS_CC2'].median(), inplace=True)
df['LDAPS_CC3'].fillna(df['LDAPS_CC3'].median(), inplace=True)
df['LDAPS_CC4'].fillna(df['LDAPS_CC4'].median(), inplace=True)
df['LDAPS_PPT1'].fillna(df['LDAPS_PPT1'].median(), inplace=True)
df['LDAPS_PPT2'].fillna(df['LDAPS_PPT2'].median(), inplace=True)
df['LDAPS_PPT3'].fillna(df['LDAPS_PPT3'].median(), inplace=True)
df['LDAPS_PPT4'].fillna(df['LDAPS_PPT4'].median(), inplace=True)
df['Next_Tmax'].fillna(df['Next_Tmax'].median(), inplace=True)
df['Next_Tmin'].fillna(df['Next_Tmin'].median(), inplace=True)

# Drop rows with missing values in 'station' and 'Date'
df.dropna(subset=['station', 'Date'], inplace=True)

# Verify if there are any remaining missing values
print(df.isnull().sum())
```

* It appears that there are no more missing values in any column of your dataset based on the output of print(df.isnull().sum()). This is great news because it means the dataset is now clean and ready for further analysis and modeling.

```python
# Splitting into features (X) and target variables (y)
X = df.drop(['Date', 'Next_Tmax', 'Next_Tmin'], axis=1)  # Features excluding Date and Next_Tmax/Next_Tmin
y_max = df['Next_Tmax']  # Target variable for next-day maximum temperature
y_min = df['Next_Tmin']  # Target variable for next-day minimum temperature

# Splitting into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_max_train, y_max_test = train_test_split(X, y_max, test_size=0.2, random_state=42)
_, _, y_min_train, y_min_test = train_test_split(X, y_min, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize the model
rf_max = RandomForestRegressor(random_state=42)

# Train the model
rf_max.fit(X_train, y_max_train)

# Predict on the test set
y_max_pred = rf_max.predict(X_test)

# Evaluate model performance
mse_max = mean_squared_error(y_max_test, y_max_pred)
print(f"Mean Squared Error (Next_Tmax): {mse_max}")
```

```python
# Initialize the model
rf_min = RandomForestRegressor(random_state=42)

# Train the model
rf_min.fit(X_train, y_min_train)

# Predict on the test set
y_min_pred = rf_min.predict(X_test)

# Evaluate model performance
mse_min = mean_squared_error(y_min_test, y_min_pred)
print(f"Mean Squared Error (Next_Tmin): {mse_min}")
```

```python
# Visualizing feature importance for Next_Tmax prediction
feature_importance_max = pd.Series(rf_max.feature_importances_, index=X.columns)
feature_importance_max.nlargest(10).plot(kind='barh')
plt.title('Feature Importance for Next_Tmax Prediction')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()

# Visualizing feature importance for Next_Tmin prediction
feature_importance_min = pd.Series(rf_min.feature_importances_, index=X.columns)
feature_importance_min.nlargest(10).plot(kind='barh')
plt.title('Feature Importance for Next_Tmin Prediction')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot for Next_Tmax prediction
plt.subplot(1, 2, 1)
plt.scatter(y_max_test, y_max_pred, alpha=0.5)
plt.plot([min(y_max_test), max(y_max_test)], [min(y_max_test), max(y_max_test)], 'k--', lw=2)
plt.title('Next_Tmax: Actual vs Predicted')
plt.xlabel('Actual Next_Tmax')
plt.ylabel('Predicted Next_Tmax')

# Scatter plot for Next_Tmin prediction
plt.subplot(1, 2, 2)
plt.scatter(y_min_test, y_min_pred, alpha=0.5)
plt.plot([min(y_min_test), max(y_min_test)], [min(y_min_test), max(y_min_test)], 'k--', lw=2)
plt.title('Next_Tmin: Actual vs Predicted')
plt.xlabel('Actual Next_Tmin')
plt.ylabel('Predicted Next_Tmin')

plt.tight_layout()
plt.show()
```
